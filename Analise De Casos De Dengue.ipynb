# Análise De Casos De Dengue (2016 a 2020)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dados=pd.read_csv("ph_dengue_cases2016-2020.csv",parse_dates=["Year"])

dados.head()
```

## Análise Exploratoria Dos Dados

**Vendo Algumas Informações Do DataSet**

```python
print(f"DIMENSÃO DATASET:\n ")

print(f"Linhas:{dados.shape[0]}\nColunas:{dados.shape[1]}")
```

**Vendo se temos valores duplicados**

```python
print("Quantidade De Valores Duplicados:",dados.duplicated().sum())
```

**Vendo se temos valores NAN**

```python
dados.isna().sum()
```

**Vendo a descrição Dos Dados para variaveis do tipo objeto**

```python
dados.select_dtypes("object").describe()
```

**Vendo a descrição Dos Dados para variaveis do tipo núerico**

### Visualizando Algumas Informções

**Vendo o Mês Com Mais Casos De Dengue**

```python
fig,ax=plt.subplots(figsize=(20,10))
sns.barplot(x="Month",y="Dengue_Cases",data=dados,ax=ax)
```

**Vendo o Mês Com Mais Mortes De Dengue**

```python
fig,ax=plt.subplots(figsize=(20,10))
sns.barplot(x="Month",y="Dengue_Deaths",data=dados,ax=ax)
```

**Vendo o Região Com Mais Casos De Dengue**

```python
fig,ax=plt.subplots(figsize=(20,10))
sns.barplot(x="Region",y="Dengue_Cases",data=dados,ax=ax)
```

**Vendo o Região Com Mais Mortes De Dengue**

```python
fig,ax=plt.subplots(figsize=(20,10))
sns.barplot(x="Region",y="Dengue_Deaths",data=dados,ax=ax)
```

**Número De Casos Por Ano**

```python
fig,ax=plt.subplots(figsize=(20,10))
sns.lineplot(x="Year",y="Dengue_Cases",data=dados,ax=ax)
```

**Número De Mortes Por Ano**

```python
fig,ax=plt.subplots(figsize=(20,10))
sns.lineplot(x="Year",y="Dengue_Deaths",data=dados,ax=ax)
```

**Mês e o Ano que Houve Mais casos de Dengue**

**Mês e o Ano que Houve Mais Mortes de Dengue**

```python
fig,ax=plt.subplots(figsize=(20,10))
sns.barplot(x="Month",y="Dengue_Deaths",hue="Year",data=dados,ax=ax)
```
